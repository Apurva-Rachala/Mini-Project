```python
plots={1:[111],2:[121,122],3:{131,132,133},4:[221,222,223,224],5:[231,232,233,234,235],6:[231,232,233,234,235,236]}
def boxplot(x,y,df):
    rows=int(str(plots[len(y)][0])[0])
    columns=int(str(plots[len(y)][0])[1])
    plt.figure(figsize=(7*columns,7*rows))
    
    for i, j in enumerate(y):
        plt.subplot(plots[len(y)][i])
        ax=sns.boxplot(x=x,y=j,data=df[[x,j]],palette="Blues",linewidth=1)
        ax.set_title(j)
    return plt.show()
```

### Identify fearture importance and prepare the data before the model training

```python
# unnecessary features woukd decrease the training speed, the model interpertability and the general generalizarion
#therefore, finding and selecting the most useful features in the dataset is crucial.
#assigning input features to X and output (detection a.k.a class) to y
#correlation plot for correlation of class with each of the feature gave us the features that do not have any kind of effect in the class and we remove or drop all the features that are not required.
X= df.drop(["class","likes_received","country","gender","name","posts_c","friend_count","age","userid"],axis="columns")
y=df["class"]
```

```python
#perform train/test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape # verify that the train/test split was successful
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
# Assess which feature in our traing data has the highest improtance
# randomforest classifier can enabel us with an additional output called "feature improtance"
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(X_train,y_train.values.ravel())
```

```python
#plot feature importance
feat_scores=pd.DataFrame({"Fraction of variables affected":rf.feature_importances_},index=X.columns)
feat_scores=feat_scores.sort_values(by="Fraction of variables affected") #sort dataframe based on the feature importance
feat_scores.plot(kind="barh",figsize=(10,5))
sns.despine()
```

### the above graph is generated by Random forest algorithm
### the graph indecated that "duration" tops the list of important features followed by "social".

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,confusion_matrix
model_LR=LogisticRegression()
model_LR.fit(X_train,y_train)
```

```python
y_predict=model_LR.predict(X_test)
```

```python
# precision is the ratio of TP/(TP+FP) 
# recall is the ratio of TP(TP+FN)
# F-beta score can be interpreted as a weight harmonic mean of the precision and recall
#where an F-beta score reaches its best value at 1 and worst at 0
print(classification_report(y_test,y_predict))
```

```python
#confusion matrix
cm=confusion_matrix(y_test,y_predict)
sns.heatmap(cm,annot=True)
```

### we look at the diagonal elements here and the elements "1.9e+0.2" and "88" are the number of the samples the model has correctly classified and "31" from class 1 and "5" from class 0 are the number of data points the model has mis-classified.

```python
from sklearn.metrics import roc_auc_score
```

```python
auc_score1=roc_auc_score(y_test,model_LR.predict_proba(X_test)[:,1])
auc_score2=roc_auc_score(y_test,model_svm.predict_proba(X_test)[:,1])
auc_score3=roc_auc_score(y_test,model_rf.predict_proba(X_test)[:,1])
auc_score4=roc_auc_score(y_test,model_knn.predict_proba(X_test)[:,1])
auc_score5=roc_auc_score(y_test,model_gnb.predict_proba(X_test)[:,1])
auc_score6=roc_auc_score(y_test,model_ab.predict_proba(X_test)[:,1])
```

```python
print("Logistic Regression:",auc_score1)
print("Support Vector Machine:",auc_score2)
print("Random Forest:",auc_score3)
print("K-Nearest Neighbours:",auc_score4)
print("Navie Bayes:",auc_score5)
print("Adaboost:",auc_score6)
```

```python
plt.plot(fpr1,tpr1,linestyle="--",color="orange",label="Logistic Regression")
plt.plot(fpr2,tpr2,linestyle="--",color="red",label="Support vector machine")
plt.plot(fpr3,tpr3,linestyle="--",color="green",label="Random Forest")
plt.plot(fpr4,tpr4,linestyle="--",color="yellow",label="K-Nearest Neighbour")
plt.plot(fpr5,tpr5,linestyle="--",color="blue",label="Naive Bayes")
plt.plot(fpr6,tpr6,linestyle="--",color="pink",label="Adaboost")
plt.title("reciever operator characteristics(ROC)")
plt.xlabel("false positive rate")
plt.ylabel("true positive rate")
plt.legend(loc="best")
plt.savefig("ROC",dpi=300)
plt.show()
```
